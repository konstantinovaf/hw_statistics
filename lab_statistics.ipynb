```python
# рассчитать показатели skewness и kurtosis

# skew - коэффицие́нт асимметри́и в теории вероятностей — величина, 
# характеризующая асимметрию распределения данной случайной величины: 
# положительное - выборка наклонена вправо, отрицательная - влево

# kurt - коэффицие́нт эксце́сса - мера остроты пика

from scipy.stats import t
from scipy.stats import kurtosis
from scipy.stats import skew

a = df['price']
print('Коэффициент ассиметрии skew: ', skew(a))
print('Коэффициент эксцесса kurt:', kurtosis(a))
```

```python
# Визуализировать несколько взаимосвязей между переменными
engine_size = df['engine-size']
price = df['price']
sns.regplot(x=price, y=engine_size,data=df, fit_reg=True)
plt.xlabel('Price')
plt.ylabel('Engine sizq')
plt.title('Корреляционная зависимость двух выборок')
plt.grid(False)
plt.show()
```

```python
# horsepower
# city-mpg
horsepower = df['horsepower']
city_mpg = df['city-mpg']
sns.regplot(x=city_mpg, y=horsepower,data=df, fit_reg=True)
plt.xlabel('Расход топлива в городе, city-mpg')
plt.ylabel('Лошадиные силы, horsepower')
plt.title('Корреляционная зависимость двух выборок')
plt.grid(False)
plt.show()
```

```python
# Выдвинуть несколько гипотез по набору имеющихся
# переменных и проверить эти гипотезы с использованием t-test.

# T-test используют для определения, насколько среднее исследуемой 
# даты статистически отличается от среднего значения генеральной совокупности. Вычисляется по формуле.

#  среднее значение price машин с размером двигателя больше 150 = price всей выборки
engine_size_more150 = df[df['engine-size'] > 150]['price']

stats.ttest_1samp(engine_size_more150, df['price'].mean())

# p-value равно 5.14, что намного больше стандартной альфа ошибки 0.05, поэтому мы принимаем нулевую гипотезу и можем сказать, 
# что не имеется статистическое различие между ценой на машины с размером двигателя больше 150
# и генеральной совокупности в целом.
```

```python
#Вспомогательный расчет для выдвижения гипотезы
# horsepower
# city-mpg
height = df['height']
compression_ratio = df['compression-ratio']
sns.regplot(x=compression_ratio, y=height,data=df, fit_reg=True)
plt.xlabel('compression_ratio')
plt.ylabel('height')
plt.title('Корреляционная зависимость двух выборок')
plt.grid(False)
plt.show()
```

```python
# Гипотеза: среднее значение высоты машин['height'] с ['compression-ratio']>9 
# сопоставимо с генеральной совокупностью высоты машин

compression_ratio_more9 = df[df['compression-ratio'] > 9]['height']

stats.ttest_1samp(compression_ratio_more9, df['height'].mean())
# p-value равно 0.18, что намного больше стандартной альфа ошибки 0.05, поэтому мы принимаем нулевую гипотезу и можем сказать, 
# что не имеется статистическое различие между высотой машин со степенью сжатия>9 и генеральной совокупности в целом.
```
